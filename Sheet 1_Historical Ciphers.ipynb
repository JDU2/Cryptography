```python
def caesar_encrypt(cleartext, key):
    ciphertext = "".join([chr((ord(c)+65+key)%26+65) for c in cleartext])
    return ciphertext
```

```python
caesar_encrypt("JULIAN", 8)
```

```python
def caesar_decrypt(ciphertext, key):
    cleartext = caesar_encrypt(ciphertext, (-1)*key)
    return cleartext
```

```python
caesar_decrypt('RCTQIV',8)
```

```python
def atbash(cleartext):
    E=lambda x: -(x+1)%26
    ciphertext="".join([chr(E(ord(c)-65)+65) for c in cleartext])
    return ciphertext
```

```python
atbash("JULIAN")
```

```python
atbash('QFORZM')
```

```python
def atbash_new(cleartext, key):
    alph="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    E=lambda x: -(x+1)%26
    ciphertext="".join([key[E(alph.find(c))] for c in cleartext])
    return ciphertext
```

```python
atbash_new("JULIAN","ABCDEFGHIJKLMNOPQRSTUVWXYZ")
```
